```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = '../merged'
data_path = os.path.join(path, 'all_data.csv')
full_data = pd.read_csv(data_path, low_memory=False)
```

```python
flourish = full_data['flourish'].dropna()

fig = plt.figure(figsize=(8, 5), dpi=200)
ax = fig.add_subplot(111, box_aspect=0.625)
ax.tick_params(axis='both', which='major', labelsize=30)
bins = list(np.linspace(8, 57, 50))
ax.hist(flourish, bins=bins, color='C0', edgecolor='black', density=True)
ax.set_xlabel('Flourish score', fontsize=40)
ax.set_ylabel('Percentage', fontsize=40)
ax.set_yticks(np.array([0.02,0.04,0.06,0.08]), ['2%','4%','6%','8%'])
ax.set_xticks(np.array([8,16,24,32,40,48,56]) + 0.5, ['8','16','24','32','40','48','56'])
ax.grid(False)
plt.show()
```

```python
deprawsc = full_data['deprawsc'].dropna()

fig = plt.figure(figsize=(8, 5), dpi=200)
ax = fig.add_subplot(111, box_aspect=0.625)
ax.tick_params(axis='both', which='major', labelsize=30)
bins = list(np.linspace(0, 28, 29))
ax.hist(deprawsc, bins=bins, color='C0', edgecolor='black', density=True)
ax.set_xlabel('Depression score', fontsize=40)
ax.set_ylabel('Percentage', fontsize=40)
ax.set_yticks(np.array([0.02,0.04,0.06,0.08]), ['2%','4%','6%','8%'])
ax.set_xticks(np.array([0,9,18,27]) + 0.5, ['0','9','18','27'])
ax.grid(False)
plt.show()
```

```python
anx_score = full_data['anx_score'].dropna()

fig = plt.figure(figsize=(8, 5), dpi=200)
ax = fig.add_subplot(111, box_aspect=0.625)
ax.tick_params(axis='both', which='major', labelsize=30)
bins = list(np.linspace(0, 21, 22))
ax.hist(anx_score, bins=bins, color='C0', edgecolor='black', density=True)
ax.set_xlabel('Anxiety score', fontsize=40)
ax.set_ylabel('Percentage', fontsize=40)
ax.set_yticks(np.array([0.02,0.04,0.06,0.08]), ['2%','4%','6%','8%'])
ax.set_xticks(np.array([0,7,14,21]) + 0.5, ['0','7','14','21'])
ax.grid(False)
plt.show()
```
