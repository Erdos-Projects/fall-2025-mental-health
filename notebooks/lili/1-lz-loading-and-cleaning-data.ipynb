```python
import numpy as np
import pandas as pd

FILE_NAMES = [
    '../../data/raw/HMS_2024-2025_PUBLIC_instchars.csv',
    '../../data/raw/HMS_2023-2024_PUBLIC_instchars.csv',
    '../../data/raw/HMS_2022-2023_PUBLIC_instchars.csv',
    '../../data/raw/HMS_2021-2022_PUBLIC_instchars.csv',
    '../../data/raw/HMS_2019-2020_PUBLIC_instchars.csv'
    ]
```

```python
NEW_FILE_NAMES = [
    '../../data/interim/2024-2025_lili.csv',
    '../../data/interim/2023-2024_lili.csv',
    '../../data/interim/2022-2023_lili.csv',
    '../../data/interim/2021-2022_lili.csv',
    '../../data/interim/2019-2020_lili.csv'
]

for file_name in NEW_FILE_NAMES:
    df_empty = pd.DataFrame(columns=['responseid', 'will_remove', 'fincur', 'finpast', 'food_worry', 'food_notlast', 'housing_worry', 'dependents', 'educ_par1', 'educ_par2', 'educ_par1_rel', 'educ_par2_rel', 'living_sit', 'residenc'])
    df_empty.to_csv(file_name, index=False)
```

```python
for i in range(5):
    print(f'reading {FILE_NAMES[i]}')
    df = pd.read_csv(FILE_NAMES[i])
    df_new = pd.read_csv(NEW_FILE_NAMES[i])
    df_new['responseid'] = df['responseid']
    df_new['will_remove'] = 0

    df_new['fincur'] = df['fincur']
    #df_new.loc[df_new['fincur'].isna(), 'will_remove'] = 1

    df_new['finpast'] = df['finpast']
    #df_new.loc[df_new['finpast'].isna(), 'will_remove'] = 1

    df_new['food_worry'] = df['food_worry']
    #df_new.loc[df_new['food_worry'].isna(), 'will_remove'] = 1

    df_new['food_notlast'] = df['food_notlast']

    if i < 3:
        df_new['housing_worry'] = df['housing_worry']
        #df_new.loc[df_new['housing_worry'].isna(), 'will_remove'] = 1

    if i < 2:
        df_new['dependents'] = df['dependents']

    if i >= 3:
        df_new['dependents'] = df['children_dep']
        #df_new.loc[df_new['children_dep'].isna(), 'will_remove'] = 1

    df_new['educ_par1'] = df['educ_par1']
    df_new.loc[df['educ_par1'] == 8 , 'educ_par1'] = np.nan
    #df_new.loc[df_new['educ_par1'].isna(), 'will_remove'] = 1

    df_new['educ_par2'] = df['educ_par2']
    df_new.loc[df['educ_par2'] == 8 , 'educ_par2'] = np.nan
    #df_new.loc[df_new['educ_par2'].isna(), 'will_remove'] = 1

    df_new['educ_par1_rel'] = df['educ_par1_rel']
    df_new.loc[df['educ_par1_rel'] == 3 , 'educ_par1'] = np.nan
    df_new.loc[df['educ_par1_rel'] == 3 , 'educ_par1_rel'] = np.nan

    df_new['educ_par2_rel'] = df['educ_par2_rel']
    df_new.loc[df['educ_par2_rel'] == 3 , 'educ_par2'] = np.nan
    df_new.loc[df['educ_par2_rel'] == 3 , 'educ_par2_rel'] = np.nan

    if i < 2:
        df_new['living_sit'] = df['living_sit']
        df_new.loc[df['living_sit'] == 4 , 'living_sit'] = np.nan

    df_new['residenc'] = df['residenc']
    df_new.loc[df['residenc'] == 7 , 'residenc'] = np.nan

    
    df_new.to_csv(NEW_FILE_NAMES[i], index=False)
```

```python
#testing the saved files

for file_name in NEW_FILE_NAMES:
    print(f'reading {file_name}')
    df = pd.read_csv(file_name)
    print(df.dependents.fillna(-1).value_counts())
    print('-'*80)
```
